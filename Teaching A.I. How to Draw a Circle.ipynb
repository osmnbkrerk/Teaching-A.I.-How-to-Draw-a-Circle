```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from matplotlib import pyplot as plt
import numpy as np
import math 
```

```python
angle = np.linspace( 0 , 2 * 3.14 , 150 ) 
 
radius = 0.4
 
X = radius * np.cos( angle ) 
y = radius * np.sin( angle ) 

X = X[:75]
y = y[:75]

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, shuffle = True)
```

```python
# we define the keras model
model = Sequential([
    Dense(units=1500, input_dim=1, activation='relu'),
    Dense(units=1500,  activation='relu'),
    Dense(units=1500,  activation='relu'),
    Dense(units=1, activation='linear')
])

# compile the keras model
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error', metrics=['accuracy'])

# fit the keras model on the dataset (CPU)
model.fit(x_train, y_train, epochs=1000, shuffle=True)
```

```python
# make class predictions with the model
X = radius * np.cos( angle ) 
predictions = model.predict(X)

y_bottom = predictions * -1
predictions = np.append(predictions, y_bottom)
```

```python
y = radius * np.sin( angle ) 
x_bottom = np.append(X,X * -1)
```
